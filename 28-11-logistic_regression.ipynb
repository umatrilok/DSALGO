```python
import pandas as pd
```

```python
fruits = pd.read_csv('fruits.csv')
```

```python
X = fruits[['sphericity','weight']]
y = fruits['label']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X,y)
```

```python
b,w = model.intercept_,model.coef_
w,b
```

```python
#list of list we are getting, we have to get that perticular value only---> so do model.coef_[0]
```

```python
b,w = model.intercept_,model.coef_[0]
w,b
```

```python
y1 = X.dot(w)+b     # there is a conflict with y so changing to y1
y1  ### We do not need y----------------> we need p--> PROBABILITY
```

```python
y1 = X.dot(w)+b 
p = 1/(1+np.exp(-y1))  #PROBABILITY
```

```python
p,1-p
```

```python
np.array(['ORANGE' if i>= 0.5 else 'APPLE' for i in p])
```

```python
# we can make the model to predict as well 
```

```python
model.predict(X)
```

```python
model.predict_proba(X)
```

```python
model.predict_proba(X)[:10]  # IF WE ADD BOTH THE PROBABILITY IT IS GOING TO BE 1 
```
